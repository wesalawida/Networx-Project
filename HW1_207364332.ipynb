# Hw 1– Networkx and Network Measures

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

nodes = pd.read_csv('GOT_nodes.csv') 
edges = pd.read_csv('GOT_edges.csv')
```

# *Question 3 :*

```python
Graphtype = nx.Graph()
G = nx.from_pandas_edgelist(edges, source='source', target='target', edge_attr='weight')
from matplotlib.pyplot import figure
figure(figsize=(30, 50))
pos = nx.spring_layout(G, k=50)  # For better example looking
nx.draw(G, pos, with_labels=True,node_size=4000,font_size=20,node_color="red",edge_color="blue" )
labels = {e: G.edges[e]['weight'] for e in G.edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels,font_size=20)
plt.show()
```

# *Question 5 :*

```python
sp = dict(nx.all_pairs_dijkstra_path_length(G))
```

```python
max(sp[0])
all_values = sp[0].values()
max_value = max(all_values)
```

```python
res = {key: max(val.values()) for key, val in sp.items()}
```

```python
res
```

```python
itemMaxValue = max(res.items(), key=lambda x: x[1])

listOfKeys = list()
for key, value in res.items():
    if value == itemMaxValue[1]:
        listOfKeys.append(key)
```

```python
itemMaxValue[1]
```

```python
names =list(nodes.iloc[listOfKeys]['name'])
```

```python
names
```

# *Question 6 :*

The first approach :

```python
a = nx.eigenvector_centrality(G,weight = 'weights')
```

```python
a = sorted(a.items(), key=lambda item: item[1],reverse=True)[:5]
a
```

```python
list(nodes.iloc[list(list(zip(*a))[0])]['name'])
```

```python
b=nx.betweenness_centrality(G,weight='weight', normalized=False)
```

```python
b = sorted(b.items(), key=lambda item: item[1],reverse=True)[:5]
b
```

```python
list(nodes.iloc[list(list(zip(*b))[0])]['name'])
```

# *Question 7 :*

```python
G1=nx.erdos_renyi_graph(107,0.1,seed=None, directed=False)
# the average path length
avg1 = nx.average_shortest_path_length(G1)
# clustering coefficient 
clus1 = nx.average_clustering(G1)

print("The average path length of the graph: ",avg1)
print("The clustering coefficient of the graph: ",clus1)

# degree distribution
degrees1 = [d for n, d in G1.degree()]
plt.hist(degrees1)

```
